```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr

# ----------------------
# 1. 데이터 수집
# ----------------------
start = '2018-01-01'
end = '2026-03-31'

# 금리 (FRED)
y10 = pdr.DataReader('DGS10', 'fred', start, end)  # 10Y
y2 = pdr.DataReader('DGS2', 'fred', start, end)    # 2Y

# 시장 데이터 (Yahoo)

# ❌ 제거
# yf.pdr_override()

# ✅ 그대로 사용
bank = yf.download('KBE', start=start, end=end)['Close']
vix = yf.download('^VIX', start=start, end=end)['Close']

# ----------------------
# 2. 데이터 정리
# ----------------------
df = pd.concat([y10, y2, bank, vix], axis=1)
df.columns = ['y10', 'y2', 'bank', 'vix']
df = df.dropna()

# 수익률
df['bank_return'] = df['bank'].pct_change()

# ----------------------
# 3. Stress Index 계산
# ----------------------

# Spread
df['spread'] = df['y10'] - df['y2']
df['spread_change'] = df['spread'].diff()

# 방향 맞추기
df['rate_stress'] = df['spread_change']
df['market_stress'] = -df['bank_return']
df['vix_stress'] = df['vix']

# Z-score 함수
def zscore(series, window=60):
    return (series - series.rolling(window).mean()) / series.rolling(window).std()

df['rate_z'] = zscore(df['rate_stress'])
df['market_z'] = zscore(df['market_stress'])
df['vix_z'] = zscore(df['vix_stress'])

# 최종 Stress Index
df['stress_index'] = (
    0.3 * df['rate_z'] +
    0.4 * df['market_z'] +
    0.3 * df['vix_z']
)

# smoothing
df['stress_index_smooth'] = df['stress_index'].rolling(5).mean()
df['decay_stress'] = df['stress_index'].ewm(span=20).mean()
# ----------------------
# 4. 시각화
# ----------------------
plt.figure(figsize=(12,6))
plt.plot(df.index, df['stress_index_smooth'], label='Stress Index')

# 위험 구간 표시
plt.axhline(1.5, linestyle='--', label='Risk Threshold')
plt.axhline(2.0, linestyle='--', label='Crisis Threshold')

#plt.axhline(0.5, linestyle='--', label='Risk Threshold')
#plt.axhline(1.0, linestyle='--', label='Crisis Threshold')

plt.title('Bank Stress Index')
plt.legend()
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr

# ----------------------
# 1. 데이터 수집
# ----------------------
start = '2018-01-01'
end = '2024-12-31'

# 금리 (FRED)
y10 = pdr.DataReader('DGS10', 'fred', start, end)  # 10Y
y2 = pdr.DataReader('DGS2', 'fred', start, end)    # 2Y

# 시장 데이터 (Yahoo)

# ❌ 제거
# yf.pdr_override()

# ✅ 그대로 사용
bank = yf.download('KBE', start=start, end=end)['Close']
vix = yf.download('^VIX', start=start, end=end)['Close']

# ----------------------
# 2. 데이터 정리
# ----------------------
df = pd.concat([y10, y2, bank, vix], axis=1)
df.columns = ['y10', 'y2', 'bank', 'vix']
df = df.dropna()

# 수익률
df['bank_return'] = df['bank'].pct_change()

# ----------------------
# 3. Stress Index 계산
# ----------------------

# Spread
df['spread'] = df['y10'] - df['y2']
df['spread_change'] = df['spread'].diff()

# 방향 맞추기
df['rate_stress'] = df['spread_change']
df['market_stress'] = -df['bank_return']
df['vix_stress'] = df['vix']

# Z-score 함수
def zscore(series, window=60):
    return (series - series.rolling(window).mean()) / series.rolling(window).std()

df['rate_z'] = zscore(df['rate_stress'])
df['market_z'] = zscore(df['market_stress'])
df['vix_z'] = zscore(df['vix_stress'])

# 최종 Stress Index
df['stress_index'] = (
    0.3 * df['rate_z'] +
    0.4 * df['market_z'] +
    0.3 * df['vix_z']
)

# smoothing
df['stress_index_smooth'] = df['stress_index'].rolling(5).mean()
df['decay_stress'] = df['stress_index'].ewm(span=20).mean()
# 1. decay smoothing
df['decay_smooth'] = df['decay_stress'].rolling(3).mean()

# 2. hysteresis
upper = 0.6
lower = 0.4
# ----------------------
# 4. 시각화
# ----------------------
plt.figure(figsize=(12,6))
plt.plot(df.index, df['decay_stress'], label='Stress Index')

# 위험 구간 표시
#plt.axhline(1.5, linestyle='--', label='Risk Threshold')
#plt.axhline(2.0, linestyle='--', label='Crisis Threshold')

plt.axhline(0.5, linestyle='--', label='Risk Threshold')
plt.axhline(1.0, linestyle='--', label='Crisis Threshold')

plt.title('Bank Stress Index')
plt.legend()
plt.grid()
plt.show()
```

```python
upper = 0.6
lower = 0.4

state = False
states = []

for x in df['decay_stress']:
    if not state and x > upper:
        state = True
    elif state and x < lower:
        state = False
    states.append(state)

df['aftershock'] = states
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['aftershock'], label='aftershock')
```
